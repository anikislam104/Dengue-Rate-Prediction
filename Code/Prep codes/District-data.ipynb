```python
import pandas as pd

def merge_sheets(excel_file_path):
    # Read the Excel file
    excel_data = pd.read_excel(excel_file_path, sheet_name=None)
    
    # Get the names of the first and second sheets
    sheet_names = list(excel_data.keys())
    sheet1_name = sheet_names[0]
    sheet2_name = sheet_names[1]
    
    # Get the data from the first and second sheets
    sheet1_data = excel_data[sheet1_name]
    sheet2_data = excel_data[sheet2_name]
    
    # Merge the data from the first and second sheets
    merged_data = pd.concat([sheet1_data, sheet2_data], axis=0)
    
    # Create a new Excel writer object
    writer = pd.ExcelWriter(excel_file_path, engine='openpyxl')
    
    # Write the merged data to a new sheet
    merged_data.to_excel(writer, sheet_name='Merged', index=False)
    
    # Save the changes to the Excel file
    writer.save()
    writer.close()

# for all the files in the directory run the merge_sheets function
for file in os.listdir('../Dataset/Hospital/2022-11 ok/'):
    if file.endswith(".xlsx"):
        merge_sheets(os.path.join('../Dataset/Hospital/2022-11 ok/', file))
```

```python
def keep_data_tables(filePath,sheetNames):
    workbook = openpyxl.load_workbook(filePath)    
    table_names = workbook.sheetnames

    tables_to_keep = sheetNames
    for table_name in table_names:
        if table_name not in tables_to_keep:
            del workbook[table_name]

    
    workbook.save(filePath)    
    workbook.close()
```

```python
import openpyxl

def delete_rows(excel_file_path, sheet1_rows_to_delete, sheet2_rows_to_delete):
    workbook = openpyxl.load_workbook(excel_file_path)
    sheets = workbook.sheetnames
    
    # delete rows from first sheet
    sheet1 = workbook[sheets[0]]
    for row in sheet1_rows_to_delete:
        sheet1.delete_rows(row)
    
    # delete rows from second sheet
    sheet2 = workbook[sheets[1]]
    for row in sheet2_rows_to_delete:
        sheet2.delete_rows(row)
    
    workbook.save(excel_file_path)
    workbook.close()
```

```python
import openpyxl
import re

def convert_bengali_to_english_numerals(excel_file_path):
    # load the workbook
    workbook = openpyxl.load_workbook(excel_file_path)

    # loop through each sheet in the workbook
    for sheet in workbook:
        # loop through each cell in the sheet
        for row in sheet.iter_rows():
            for cell in row:
                # check if the cell contains Bengali numerals
                if re.search('[০-৯]', str(cell.value)):
                    # replace Bengali numerals with English numerals
                    cell.value = re.sub('[০-৯]', lambda x: str(ord(x.group(0)) - 2534), str(cell.value))

    # save the updated workbook
    workbook.save(excel_file_path)
    workbook.close()

```

```python
import openpyxl

def delete_columns_and_rows(excel_file_path, sheet1_columns_to_delete, sheet1_rows_to_delete, sheet2_columns_to_delete, sheet2_rows_to_delete):
    workbook = openpyxl.load_workbook(excel_file_path)
    sheets = workbook.sheetnames
    
    # delete columns and rows from first sheet
    sheet1 = workbook[sheets[0]]
    for column in sheet1_columns_to_delete:
        sheet1.delete_cols(column)
    for row in sheet1_rows_to_delete:
        sheet1.delete_rows(row)
    
    # delete columns and rows from second sheet
    sheet2 = workbook[sheets[1]]
    for column in sheet2_columns_to_delete:
        sheet2.delete_cols(column)
    for row in sheet2_rows_to_delete:
        sheet2.delete_rows(row)
    
    workbook.save(excel_file_path)
    workbook.close()
```

```python
# add a row on top of each sheet with every cell containing the letter "A" and row length equal to the number of columns in the sheet
import os
import glob
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter

def add_row_on_top(directory_path):
# Get all Excel files in the directory
    excel_files = glob.glob(directory_path + '/*.xlsx')

    # Loop through each Excel file
    for excel_file in excel_files:
        # Load the workbook
        workbook = load_workbook(excel_file)
        
        # Loop through each sheet in the workbook
        for sheet_name in workbook.sheetnames:
            # Get the sheet
            sheet = workbook[sheet_name]
            
            # Get the number of columns in the sheet
            number_of_columns = sheet.max_column
            
            # Create a list of "A" with length equal to the number of columns
            row_data = ["A"] * number_of_columns
            
            # Add the row at the top of the sheet
            sheet.insert_rows(1)
            
            # Loop through each cell in the row
            for column, cell in enumerate(sheet[1]):
                # Assign the value from the list to the cell
                cell.value = row_data[column]
        
        # Save the modified workbook
        workbook.save(excel_file)
        # print(excel_file + " completed")

directory_path = 'test_dir/test-district/'
for subdirectory in os.listdir(directory_path):
    if os.path.isdir(os.path.join(directory_path, subdirectory)):
        add_row_on_top(os.path.join(directory_path, subdirectory))
        print(subdirectory + " completed")

```

```python
# combine all sheets into one sheet
import pandas as pd

def merge_excel_sheets(file_path):
    print(file_path + " started")
    # Read both sheets into dataframes
    sheets = pd.read_excel(file_path, sheet_name=None)


    # Get the names of all sheets win another way
    sheet_names = list(sheets.keys())
    


    # Assuming the first sheet is the reference sheet
    reference_sheet_name = sheet_names[0]
    reference_sheet = sheets[reference_sheet_name]

    # Initialize a new dataframe for merged data
    merged_df = pd.DataFrame()

    # Iterate through columns in the reference sheet
    for column in reference_sheet.columns:
        # Concatenate corresponding columns from all sheets
        merged_column = pd.concat([sheets[sheet_name][column] for sheet_name in sheet_names], ignore_index=True)
        # Add the concatenated column to the merged dataframe
        merged_df[column] = merged_column

    # Write the merged data to a new sheet
    merged_df.to_excel(file_path, sheet_name='Merged', index=False)

    

#directory_path

def merge_sheets(directory_path):
    excel_files = glob.glob(directory_path + '/*.xlsx')

    for excel_file in excel_files:
        merge_excel_sheets(excel_file)
        print(excel_file + " completed")

directory_path = 'test_dir/test-district/'
for subdirectory in os.listdir(directory_path):
    if os.path.isdir(os.path.join(directory_path, subdirectory)):
        merge_sheets(os.path.join(directory_path, subdirectory))
        print(subdirectory + " completed")
```

```python
from googletrans import Translator

def translate_bengali_to_english(text):
    translator = Translator(service_urls=['translate.google.com'])
    translation = translator.translate(text, src='bn', dest='en')
    return translation.text

# translate all the Bengali text in the sheet to English
import os
import glob
import pandas as pd

# Directory path
directory_path = 'test_dir/2022-11/'

# Get all Excel files in the directory
excel_files = glob.glob(directory_path + '/*.xlsx')

# Loop through each Excel file
for excel_file in excel_files:
    # Read the Excel file
    df = pd.read_excel(excel_file, sheet_name=None)
    
    # Loop through each sheet in the Excel file
    for sheet_name, sheet_df in df.items():
        # Loop through each cell in the sheet
        for row in sheet_df.iterrows():
            for cell in row[1].index:
                # Check if the cell contains Bengali text
                if type(row[1][cell]) == str and row[1][cell] != '' and ord(row[1][cell][0]) >= 2432 and ord(row[1][cell][0]) <= 2559:
                    # Translate the text
                    row[1][cell] = translate_bengali_to_english(row[1][cell])
    
    # Save the modified Excel file
    with pd.ExcelWriter(excel_file) as writer:
        for sheet_name, sheet_df in df.items():
            sheet_df.to_excel(writer, sheet_name=sheet_name, index=False)
    print(excel_file + " completed")
```

```python
import pandas as pd
from openpyxl import load_workbook

def delete_rows_from_sheets(file_path):
    # Read all sheets into a dictionary
    sheets = pd.read_excel(file_path, sheet_name=None)

    # Iterate through each sheet
    for sheet_name, sheet_df in sheets.items():
        # Check if it's the first sheet
        if sheet_name == list(sheets.keys())[0]:
            # Delete the first 4 rows from the first sheet
            sheets[sheet_name] = sheet_df.iloc[3:]
        else:
            # Delete the first row from other sheets
            sheets[sheet_name] = sheet_df.iloc[1:]

    # Save the modified sheets back to the Excel file
    with pd.ExcelWriter(file_path, engine='openpyxl') as writer:
        for sheet_name, sheet_df in sheets.items():
            sheet_df.to_excel(writer, sheet_name=sheet_name, index=False)

# call the function for all the files in the directory
directory_path = 'test_dir/2019-09/'
excel_files = glob.glob(directory_path + '/*.xlsx')

for excel_file in excel_files:
    delete_rows_from_sheets(excel_file)
    print(excel_file + " completed")
```
