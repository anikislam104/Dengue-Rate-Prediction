```python
import pandas as pd
import numpy as np
import openpyxl
import os
import glob
```

```python
dfs = []
filenames = []

dir =  "../../Data/Processed Data/Districts Dataset"

for file in os.listdir(dir):
    if file.endswith(".xlsx"):
        df = pd.read_excel(dir + "/" + file)
        print("Reading file: ", file, " at index: ", len(dfs))
        dfs.append(df)
        filenames.append(file)
```

```python
for df in dfs:
    print(df.shape)
```

```python
# Get the common dates from 'Date'column meaning dates that are available in all the dataframes
common_dates = set(dfs[0]['Date'])
for df in dfs:
    common_dates = common_dates.intersection(set(df['Date']))

common_dates = list(common_dates)
common_dates.sort()

common_dates
```

```python
# from each dataframe, only keep the rows that have the common dates
for i in range(len(dfs)):
    dfs[i] = dfs[i][dfs[i]['Date'].isin(common_dates)]

# now print the shape of each dataframe
for df in dfs:
    print(df.shape)
```

```python
# if in any dataframe there are any Date duplicates, then merge them as one enntry by taking the sum of the values

for i in range(len(dfs)):
    dfs[i] = dfs[i].groupby('Date').sum().reset_index()

# now print the shape of each dataframe
for df in dfs:
    print(df.shape)
```

```python
# save each dataframe to separate excel files with the same name as the original file
dir = "../../Data/Processed Data/Districts Dataset/More Processed"

for i in range(len(dfs)):
    dfs[i].to_excel(dir + "/" + filenames[i], index=False)
```
