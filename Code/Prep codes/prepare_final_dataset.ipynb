```python
import pandas as pd
import numpy as np
import openpyxl
import os
import glob
```

```python
dir = "../../Data/Processed Data/test_read"

dfs = []
filenames = []

for file in glob.glob(os.path.join(dir, "*.xlsx")):
    df = pd.read_excel(file)
    dfs.append(df)
    #removing the extension from the filename
    filenames.append(os.path.basename(file).split(".")[0])
```

```python
dfs[0].head()
```

```python
#print the unique shapes of the dataframes 
print(set([df.shape for df in dfs]))
```

```python
# includinf all the dataframes, find out how many common date values are there in the dataframes
common_dates = set(dfs[0].Date)
for df in dfs[1:]:
    common_dates = common_dates.intersection(df.Date)

print(len(common_dates))
```

```python
#filter the dataframes to only include the common dates
for i, df in enumerate(dfs):
    dfs[i] = df[df.Date.isin(common_dates)]
```

```python
#print the unique shapes of the dataframes 
print(set([df.shape for df in dfs]))
```

```python
dir = "../../Data/Processed Data/test_write"

#write the dataframes to excel files with the same filenames
for i, df in enumerate(dfs):
    df.to_excel(os.path.join(dir, filenames[i] + ".xlsx"), index=False)
    print(filenames[i] + ".xlsx is written")
```

```python
# Initialize an empty numpy array with the required shape
timestep = dfs[0].shape[0]
hospitals = len(dfs)
affected_array = np.zeros((timestep, hospitals, 5))

# Populate the numpy array with the 2nd to last column values from each dataframe
for i in range(hospitals):
    print(filenames[i])
    affected_array[:, i, :] = dfs[i].iloc[:, 1:-1].values

# Save the numpy array to a .npy file
np.save('../../Data/BD_dengue_and_weather.npy', affected_array)

# split affected_array into training,val,test (75%,15%,10%)

train_size = int(affected_array.shape[0] * 0.75)
val_size = int(affected_array.shape[0] * 0.15)
test_size = int(affected_array.shape[0] * 0.10)

train_data = affected_array[:train_size]
val_data = affected_array[train_size:train_size+val_size]
test_data = affected_array[train_size+val_size:]

np.save('../../Data/BD_dengue_and_weather_train.npy', train_data)
np.save('../../Data/BD_dengue_and_weather_val.npy', val_data)
np.save('../../Data/BD_dengue_and_weather_test.npy', test_data)
```

```python
dfs = []

dir = "../../Data/Processed Data/BD Dataset with weather"
filenames = []

for file in glob.glob(os.path.join(dir, "*.xlsx")):
    df = pd.read_excel(file)
    dfs.append(df)
    #removing the extension from the filename
    filenames.append(os.path.basename(file).split(".")[0])
```

```python
dfs[0].head()
```

```python
timesteps = dfs[0].shape[0]
timesteps
```

```python
nodes = len(dfs)
nodes
```

```python
features = dfs[0].shape[1] - 1
features
```

```python
# now create a numpy array with the shape (timesteps, nodes, features)
# features = all columns except the first one (date)
data = np.zeros((timesteps, nodes, features))

for i, df in enumerate(dfs):
    data[:, i, :] = df.iloc[:, 1:].values

data.shape
```

```python
# save it in a .npy file
np.save('../../Data/BD_dengue_and_weather.npy', data)
```

```python
# split into training, validation and test sets
train_size = int(data.shape[0] * 0.75)
val_size = int(data.shape[0] * 0.15)
test_size = int(data.shape[0] * 0.10)

train_data = data[:train_size]
val_data = data[train_size:train_size+val_size]
test_data = data[train_size+val_size:]

np.save('../../Data/BD_dengue_and_weather_train.npy', train_data)
np.save('../../Data/BD_dengue_and_weather_val.npy', val_data)
np.save('../../Data/BD_dengue_and_weather_test.npy', test_data)
```

```python
# Add a column Coordinates to the curr_coords dataframe
# If the coordinates are present in the prev_coords dataframe, then add the corresponding coordinates matching 'Hospital' columns
# If the coordinates are not present in the prev_coords dataframe, then add the coordinates as 'latitude, longitude'
# Save the curr_coords dataframe to a new excel file

curr_coords['Coordinates'] = curr_coords['Hospital'].apply(lambda x: prev_coords[prev_coords['Hospital'] == x]['Coordinates'].values[0] if x in prev_coords['Hospital'].values else 'latitude, longitude')

curr_coords.to_excel("../../Data/coordinates.xlsx", index=False)
```
